```python
def max_num_post(soup):
    search_number = soup.find('em', {'class': 'search_number'}).text
    print(search_number)
    numbers = re.sub(r'[^\d,]', '', search_number)
    return int(numbers.replace(',', ''))

base_url = "https://section.blog.naver.com/Search/Post.naver"
params = {
    "pageNo": 1,
    "rangeType": "PERIOD",
    "orderBy": "sim",
    "startDate": "2020-01-01",
    "endDate": "2024-03-07",
    "keyword": f"{city} ì—¬í–‰ ì¼ê¸°"
}

# ì¿¼ë¦¬ íŒŒë¼ë¯¸í„°ë¥¼ URL ì¸ì½”ë”©
encoded_params = urlencode(params, quote_via=quote_plus)

# ìµœì¢… URL ìƒì„±
search_url = f"{base_url}?{encoded_params}"

driver.get(search_url)
# ì „ì²´ í˜ì´ì§€ ìˆ˜ íŒŒì•…
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
max_num = max_num_post(soup)
# max_num = int(max_num*0.2) #ìœ ëª… ì—¬í–‰ì§€ì¼ ê²½ìš°(ì •í™•ë„ìˆœ ì •ë ¬ì´ê¸°ë•Œë¬¸ì— ì¼ë¶€ë§Œ ì‚¬ìš©)
max_num_page = max_num//7
print(max_num_page)
```

```python

for i in range(1, max_num_page+1):
#     for i in range(1, 5):
    base_url = "https://section.blog.naver.com/Search/Post.naver"
    params = {
        "pageNo": i,
        "rangeType": "PERIOD",
        "orderBy": "sim",
        "startDate": "2020-01-01",
        "endDate": "2024-02-07",
        "keyword": f"{city} ì—¬í–‰ ì¼ê¸°"
    }

    # ì¿¼ë¦¬ íŒŒë¼ë¯¸í„°ë¥¼ URL ì¸ì½”ë”©
    encoded_params = urlencode(params, quote_via=quote_plus)

    # ìµœì¢… URL ìƒì„±
    search_url = f"{base_url}?{encoded_params}"

    driver.get(search_url)
    time.sleep(5)  
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    
    navigation = driver.find_element(By.CLASS_NAME, "pagination")  # í´ë˜ìŠ¤ ì´ë¦„ í™•ì¸ í•„ìš”
    pages = navigation.find_elements(By.TAG_NAME, "a")

    results = driver.find_elements(By.CLASS_NAME, "desc_inner")
    k = 0
    time.sleep(5)
    for result in results:
        try:
            result.click()
            time.sleep(5)  
            windows = driver.window_handles
            driver.switch_to.window(windows[-1])

            iframe = driver.find_element(By.TAG_NAME, 'iframe')
            driver.switch_to.frame(iframe)
            time.sleep(5)
            
            new_page_source = driver.page_source
            new_soup = BeautifulSoup(new_page_source, 'html.parser')

            title_element = new_soup.find('div', class_="pcol1")
            title = title_element.get_text().strip()

            date_element = new_soup.find('span', class_="se_publishDate pcol2")
            publishDate = date_element.get_text().strip()

            text_elements = new_soup.find_all('div', class_="se-module se-module-text")

            all_text = ""  # ëª¨ë“  í…ìŠ¤íŠ¸ë¥¼ ì €ì¥í•  ë¹ˆ ë¬¸ìì—´ ì´ˆê¸°í™”
            for element in text_elements:
                all_text += element.get_text().strip() + "\n"  # ê° ìš”ì†Œì˜ í…ìŠ¤íŠ¸ ì¶”ì¶œ í›„ ê°œí–‰ ë¬¸ìì™€ í•¨ê»˜ ì¶”ê°€
            if len(all_text) == 0:
                print("No content")
                driver.close()  
                driver.switch_to.window(windows[0]) 
                continue
            else:
                meta_tag = new_soup.find('meta', property='og:url')
                url = meta_tag['content']

                data.append({
                "URL": url,
                "title": title,
                "text": all_text,
                "city": city,
                "country": country,
                "publish date": publishDate
                })

                k+=1
                print("page:", i, "post:", k, "title:", title, "publish date:", publishDate)
                time.sleep(5)

                driver.close()  
                driver.switch_to.window(windows[0]) 

                time.sleep(5)
        except:
            print("Fail, move on the next")
time.sleep(5)
```

```python
def remove_illegal_chars_for_excel(value):
    # Excelì— í—ˆìš©ë˜ì§€ ì•ŠëŠ” ë¬¸ì ëª©ë¡
    illegal_chars = [chr(i) for i in range(32) if i not in [9, 10, 13]]  # ì œì–´ ë¬¸ì ì œê±°, íƒ­(9), ì¤„ë°”ê¿ˆ(10, 13)ì€ í—ˆìš©
    if isinstance(value, str):
        for char in illegal_chars:
            value = value.replace(char, '')  # í—ˆìš©ë˜ì§€ ì•ŠëŠ” ë¬¸ì ì œê±°
    return value

df = df.applymap(remove_illegal_chars_for_excel)

# Excel íŒŒì¼ë¡œ ì €ì¥ # _max
df.to_excel(f'data/{country}/{country}({city})_max.xlsx', index=False, engine='openpyxl')
```
